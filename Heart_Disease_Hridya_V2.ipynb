```python
# Heart Diseasese Data set - Done by Hridya 
# Completed on 05/18/2023

import pandas as pd
import matplotlib.pyplot as plt

# Load the heart disease data set from github
heart_df=pd.read_csv("https://raw.githubusercontent.com/hridyasg/Data_Analytics/main/Heart_Disease_Dataset.csv?classId=fa8be78a-1b27-4fb4-9128-5eddcd42f0aa&assignmentId=76d310f7-aa14-4426-84fc-ed565fa70432&submissionId=302c7c5f-7615-99fe-bb20-0c394a4bf2e0")

#column name and info
#heart_df.columns
#heart_df.info()
#heart_df

# Remove duplicates 
heart_df.drop_duplicates(inplace=True)


# Replacing sex=0 as "Female" and sex=1 as "Male"
heart_df["sex"]=heart_df["sex"].replace(to_replace=[1,0],value=["Male","Female"])
heart_df["sex"].astype(str)

# Check any null values
print(heart_df.isnull().values.any())

##############################################

# Calculate the statistics of applicable columns
print(heart_df.describe())

####################################################

# HISTOGRAM
# Histogram - frequency distibution of heart disease according to Age

hrt_dise_present= heart_df.loc[heart_df["target"]==1,"age"]                              
n,bin,patches=plt.hist(hrt_dise_present,bins=10,range=[30,80],label= "Age",ec="black")
plt.title("Heart Disease - Age Frequency")
plt.xlabel("Age")
plt.ylabel("Frequency- Heart Disease Present")
plt.legend(title="Target Vs Age")

# Label X axis age values from 30 - 80 
plt.xticks([30,35,40,45,50,55,60,65,70,75,80])
plt.show()
print("\n")

######################################################

# SCATTER PLOT 
# Calculate the correlation between Cholestrol and Resting blood pressure of individuals have chances of heart disease(Target =1)

fig,axs=plt.subplots()
x=heart_df["age"]
y=heart_df["thalach"]
axs.scatter(x,y)
axs.set_title("Age Vs Exercise Heart Rate")
axs.set_xlabel("Age")
axs.set_ylabel("Exercise heart Rate (bpm)")
plt.show()
print("\n")
corr1=round(x.corr(y),3)
print("Correlation between Age Vs Exercise Heart Rate=",corr1)
print("\n")
########################################################

# BOX PLOT
# The chest pain type in Male and Female 

pd.plotting.boxplot(heart_df,column=["trestbps"],by=["sex"],grid=True)
plt.ylabel("Trestbps (mmHg)")
plt.show()

########################################################


```

```python
##############################################
# Bar plot to show the bloood pressure for men and women of different ages 
# Blood pressure is same in young men and women 
# BLood pressure increases in women after menopause


age_filt=heart_df["age"] > 45
age_old=heart_df[age_filt]

age_less=heart_df["age"] <= 45
age_young=heart_df[age_less]

fig, axs1=plt.subplots()
df_bphigh=age_old.groupby("sex")["trestbps"].mean()
df_bphigh.plot( kind="bar",width=.25,stacked=True,legend=True,label="BP (age > 45)")
df_bplow=age_young.groupby("sex")["trestbps"].mean()
df_bplow.plot(kind="bar",width=.25,color="orange",legend=True,label="BP (age <=45)")
axs1.legend(loc="upper center")
axs1.set_title("Blood pressure mmHG ")
axs1.set_xlabel("Sex")
axs1.set_ylabel("Resting blood pressure mmHg")
plt.show()

##########################################################

# BAR PLOT
# BAR PLOT to show cholestrol in Male and Female

fig,axs2=plt.subplots()
filt=(heart_df['target']==1)
target1=heart_df[filt]
avg_gender=target1.groupby('sex')["chol"].mean()
labels=["Chol(mg/dL)"]
axs2=avg_gender.plot(kind='bar',title= "Cholestrol",legend=True,width=.25)
axs2.legend(labels)
axs2.set_ylabel("Avg Chol(mg/dL)")
plt.show()
print("\n")
print("\n")


###########################################################################

# Pie Chart 
fig,axs=plt.subplots()
# filt=(heart_df['target']==1)
# chest_pain_target1=heart_df[filt]
# #chest_pain_type=chest_pain_target1.groupby('cp')['target'].sum()
chest_pain_type = heart_df.groupby('cp')["target"].sum()
labels=["Asymptomatic","Atypical Angina","Non Anginal Pain","Typical Angina"]
chest_pain_type.plot(kind = 'pie',autopct='%1.1f%%',labels=labels,legend=True,title="Chest Pain Type",ylabel="",labeldistance=None)
axs.legend(loc="lower right")
plt.show()

##############################################################################

# Summary and conclusion 
# Check the individuals has the highest blood pressure , cholesterol and diabetes against Target (chances of heart disease)
print("\n")
print("Displaying 10 individuals with high cholesterol")
print("--------------------------------------------------")
chol_df=heart_df.sort_values(['chol'],ascending=False).head(10)
display(chol_df)
print("\n")

print("Displaying 10 individuals with high resting blood pressure")
print("-----------------------------------------------------------")
trest_df=heart_df.sort_values(["trestbps"],ascending=False).head(10)
display(trest_df)
print("\n")

print("Displaying 10 individuals with high fasting blood sugar")
print("---------------------------------------------------------")
fbs_df=heart_df.sort_values(['fbs'],ascending=False).head(10)
display(fbs_df)
print("\n")
```
